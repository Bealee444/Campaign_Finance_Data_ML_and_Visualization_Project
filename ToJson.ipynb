```python
import pandas as pd

# Specify the file path
algresults = "results_2000.csv"
ogfile = "indivs_Georgia20.csv"
```

```python
algdata = pd.read_csv(algresults, delimiter=',', header=0, names=['number','Contribid','hard2','soft2_0','soft2_1','hard3','soft3_0','soft3_1','soft3_2','hard6','soft6_0','soft6_1','soft6_2','soft6_3','soft6_4','soft6_5','hard8','soft8_0','soft8_1','soft8_2','soft8_3','soft8_4','soft8_5','soft8_6','soft8_7'])
algdata = algdata[algdata['Contribid'].str[0].str.islower()]




algdata = algdata[['Contribid','hard2','hard3','hard6', 'hard8']]

#algdata[['soft2_0', 'soft3_0', 'soft3_1']] = algdata[['soft2_0', 'soft3_0', 'soft3_1']].astype(float)

algdata[['hard2', 'hard3', 'hard6', 'hard8']] = algdata[['hard2', 'hard3', 'hard6', 'hard8']].astype(int)

print(algdata.head(10))
print(len(algdata))
```

```python
# Read the CSV file into a pandas DataFrame
totalogdata = pd.read_csv(ogfile, header=0)

# Specify columns to drop
columns_to_drop = ['Cycle', 'State', 'Microfilm']

# Drop unnecessary columns
ogdata = totalogdata.drop(columns=columns_to_drop)

# Convert 'Amount' column to float
ogdata['Amount'] = ogdata['Amount'].astype(float)

# Convert selected columns to strings
ogdata[['Contribid', 'City', 'Zip', 'Recipcode', 'Gender', 'Occupation']] = ogdata[['Contribid', 'City', 'Zip', 'Recipcode', 'Gender', 'Occupation']].astype(str)
ogdata[['Recipid','Fectransid', 'Contrib', 'Orgname', 'Ultorg', 'Realcode', 'Date', 'Street', 'Type', 'CmteID', 'OtherID',  'Employer', 'Source']] = ogdata[[ 'Recipid','Fectransid', 'Contrib', 'Orgname', 'Ultorg', 'Realcode', 'Date', 'Street', 'Type', 'CmteID', 'OtherID',  'Employer', 'Source']].astype(str)
# Display the DataFrame
print(ogdata.head())
```

```python
# Perform inner join on 'Contribid' column
merged_data = algdata.merge(ogdata, on='Contribid', how='inner')

# Drop rows where 'soft2_0' is null or NaN
merged_data = merged_data.dropna(subset=['hard2'])

# Aggregate the merged DataFrame
merged_data = merged_data.groupby('Contribid').agg({
    'Amount': 'sum',
    'hard2': 'first',   
    'hard3': 'first',
    'hard6': 'first',
    'hard8': 'first', 
    'City': 'first', 
    'Zip': 'first', 
    'Recipcode': 'first', 
    'Gender': 'first',  
    'Occupation': 'first', 
    'Recipid': 'first',
    'Fectransid': 'first',
    'Contrib': 'first',
    'Orgname': 'first',
    'Ultorg': 'first', 
    'Realcode': 'first',
    'Date': 'first',
    'Street': 'first',
    'Type': 'first',
    'CmteID':'first',
    'OtherID': 'first', 
    'Employer': 'first',
    'Source': 'first'
}).reset_index()

# Display the aggregated DataFrame
print(merged_data)
```

```python
def add_pmen_to_children(node, data):
    #print(node.get("children", []))
    for child in node.get("children", []):
        category_condition = (child["name"]) if child["name"].isdigit() else None
        if category_condition is not None:
            
            child["pmen"] = calculate_pmen(data, 'M', category_condition)
            child["punemployed"] = calculate_pemp(data, category_condition)
            child["pretired"] = calculate_pretired(data, category_condition)
            child["pdemocrat"] = calculate_pdemocrat(data, category_condition)
            child["prepublican"] = calculate_prepublican(data, category_condition)
            
            
        add_pmen_to_children(child, data)

# Assuming `dist` is your JSON structure
add_pmen_to_children(dist, merged_data)
```

```python
def calculate_pmen(data, gender_condition, category_condition):
    if len(category_condition) == 1:
        men_frequency = len(data[(data['Gender'] == gender_condition) & (data['hard2'] == int(category_condition))])
        category_frequency = len(data[data['hard2'] == int(category_condition)])
        return (men_frequency / category_frequency) * 100 if category_frequency != 0 else "wrong bea"
    
    elif len(category_condition) == 2 and category_condition[-1] != "U":
        men_frequency = len(data[(data['Gender'] == gender_condition) & (data['hard2'] == int(category_condition[0])) & (data['hard3'] == int(category_condition[1]))])
        category_frequency = len(data[(data['hard2'] == int(category_condition[0])) & (data['hard3'] == int(category_condition[1]))])
        return (men_frequency / category_frequency) * 100 if category_frequency != 0 else 0
                                      
    elif len(category_condition) == 3 and category_condition[-1] != "U":
        men_frequency = len(data[(data['Gender'] == gender_condition) & (data['hard2'] == int(category_condition[0])) & (data['hard3'] == int(category_condition[1])) & (data['hard6'] == int(category_condition[2]))])
        category_frequency = len(data[(data['hard2'] == int(category_condition[0])) & (data['hard3'] == int(category_condition[1])) & (data['hard6'] == int(category_condition[2]))])
        return (men_frequency / category_frequency) * 100 if category_frequency != 0 else 0
                                      
    elif len(category_condition) == 4 and category_condition[-1] != "U":
        men_frequency = len(data[(data['Gender'] == gender_condition) & (data['hard2'] == int(category_condition[0])) & (data['hard3'] == int(category_condition[1])) & (data['hard6'] == int(category_condition[2])) & (data['hard8'] == int(category_condition[3]))])
        category_frequency = len(data[(data['hard2'] == int(category_condition[0])) & (data['hard3'] == int(category_condition[1])) & (data['hard6'] == int(category_condition[2])) & (data['hard8'] == int(category_condition[3]))])
        return (men_frequency / category_frequency) * 100 if category_frequency != 0 else 0
                                      
    else:
        return none

```
